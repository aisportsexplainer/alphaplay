```python
df_relief_other_vars = raw_relief.copy()

df_relief_other_vars['K%'] = np.round( (df_relief_other_vars['SO'] / df_relief_other_vars['TBF']), 3)
df_relief_other_vars['BB%'] = np.round( (df_relief_other_vars['BB'] / df_relief_other_vars['TBF']), 3)

# Use FIP_IPs as has been adjusted to account for sum  up to team level (only .1 or .2 can be the decimals)
df_relief_other_vars['ERA'] = np.round( ((df_relief_other_vars['ER'] / df_relief_other_vars['IP']) * 9), 2)

# Baseball Reference has BF = AB + BB + HBP + SF 
df_relief_other_vars['BAbip'] = np.round( (df_relief_other_vars['H'] - df_relief_other_vars['HR']) / (
                                      df_relief_other_vars['TBF'] - df_relief_other_vars['BB'] -
                                      df_relief_other_vars['HBP'] - df_relief_other_vars['SO'] -
                                      df_relief_other_vars['HR']), 3)

# Baseball Reference has BF = AB + BB + HBP + SF 
df_relief_other_vars['OBP'] = np.round( (df_relief_other_vars['H'] + df_relief_other_vars['BB'] +
                                      df_relief_other_vars['HBP']) / (df_relief_other_vars['TBF']), 3)

df_relief_other_vars['SHO_per_BF'] = df_relief_other_vars['ShO'] / df_relief_other_vars['TBF']
df_relief_other_vars['CG_per_BF'] = df_relief_other_vars['CG'] / df_relief_other_vars['TBF']

# Take Tm out of the index and make it a column
df_relief_other_vars.reset_index(level=0, inplace=True)

df_relief_other_vars[['Year', 'Team', 'SO', 'BB', 'TBF', 'IP', 'FIP', 'K%', 'BB%', 'ERA', 'BAbip', 'OBP', 'SHO_per_BF', 'CG_per_BF']]

```
