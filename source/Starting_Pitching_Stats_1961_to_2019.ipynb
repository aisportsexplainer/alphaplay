```python
df_calcs = df_starters[['Year', 'tbc_id', 'retroid', 'IP']]

df_calcs['IP'] = pd.to_numeric(df_calcs['IP'], errors = 'coerce')

# Force rounding down of IP to create new count of full innings pitched (without decimal)
df_calcs['IP_v2'] = (df_calcs['IP'] // 1).astype(int)

# Separate out the decimal and divde by 3 to create the outs so when summed across rows, three outs (3*0.333) becomes 1
df_calcs['outs_v1'] = (((np.round(np.modf(df_calcs['IP'])[0], 1) * 10)).astype(int) / 3)

df_calcs['full_IPs'] = (df_calcs['IP_v2'] + df_calcs['outs_v1'])

df_calcs['Pitch_G%'] = (df_calcs['full_IPs'] / 9)

print(df_calcs.head())

df_calcs.groupby('Year').IP.mean()

df = df_calcs.groupby(['Year','tbc_id', 'retroid'])['Pitch_G%'].mean().reset_index()

df = df[['Year', 'retroid', 'Pitch_G%']]

print(df.shape)

print(df)

# Export to CSV
df.to_csv('MLB_Starters_Pitch_1961-2019.csv', header=True) 
```
